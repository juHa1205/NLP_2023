```python
from sklearn.cluster import KMeans
Kmean = KMeans(n_clusters=1)
Kmean.fit(X)

KMeans(n_clusters=1)
```

```python
X_c=Kmean.cluster_centers_
```

```python
X_c
```

```python
X_c[0]
```

```python
plt.scatter(X[:,0],X[:,1],s=50,c='b')
plt.scatter(X_c[0][0],X_c[0][1], s=200,c='g',marker='s')

plt.show()
```

```python
from sklearn.cluster import KMeans
Kmean = KMeans(n_clusters=1)
Kmean.fit(X1)

KMeans(n_clusters=1)

```

```python
X1_c=Kmean.cluster_centers_
X1_c
```

```python
plt.scatter(X1[:,0],X1[:,1],s=50,c='b')
plt.scatter(X1_c[0][0],X1_c[0][1],s=200,c='g',marker='s')

plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],s=50,c='b')
plt.scatter(X_c[0][0],X_c[0][1],s=200,c='g',marker='s')
plt.scatter(X1[:,0],X1[:,1],s=50,c='r')
plt.scatter(X1_c[0][0],X1_c[0][1],s=200,c='y',marker='s')
plt.show()
```

```python
A=np.concatenate((X,X1))
A.shape
```

```python
from sklearn.cluster import KMeans
Kmean = KMeans(n_clusters=2)
Kmean.fit(A)
```

```python
A_c=Kmean.cluster_centers_
A_c
```

```python
plt.scatter(X[:,0],X[:,1],s=50,c='b')
plt.scatter(A_c[0][0],A_c[0][1],s=200,c='g',marker='s')
plt.scatter(X1[:,0],X1[:,1],s=50,c='r')
plt.scatter(A_c[0][0],A_c[0][1],s=200,c='y',marker='s')
plt.show()
```
