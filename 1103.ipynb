```python
import numpy as np
import matplotlib.pyplot as plt

# 사각형 펄스 함수 정의
def rect(t, T, A):
    return np.where((t >= -T/2) & (t <= T/2), A, 0)

# 사각형 펄스 함수 미분 정의
def rect_derivative(t, T, A):
    return np.where((t == -T/2) | (t == T/2), A/T, 0)

# 시간 범위 설정
t = np.linspace(-5, 5, 1000)  # -5부터 5까지 1000개의 점

# 파라미터 설정
T = 2.0  # 사각형 펄스 주기
A = 1.0  # 사각형 펄스 높이

# 사각형 펄스 및 미분 계산
rect_waveform = rect(t, T, A)
rect_derivative_waveform = rect_derivative(t, T, A)

# 그래프 그리기
plt.figure(figsize=(10, 5))
plt.plot(t, rect_waveform, label='사각형 펄스')
plt.plot(t, rect_derivative_waveform, label='사각형 펄스 미분')
plt.xlabel('시간')
plt.ylabel('값')
plt.legend()
plt.grid()
plt.title('사각형 펄스 및 미분')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 사각형 펄스 함수 정의
def rect(t, T, A):
    return np.where((t >= -T/2) & (t <= T/2), A, 0)

# Fourier 변환 함수 정의
def fourier_transform(signal, time):
    return np.fft.fftshift(np.fft.fft(signal)) / len(time)

# 시간 범위 설정
t = np.linspace(-5, 5, 1000)  # -5부터 5까지 1000개의 점

# 파라미터 설정
T = 2.0  # 사각형 펄스 주기
A = 1.0  # 사각형 펄스 높이

# 사각형 펄스 계산
rect_waveform = rect(t, T, A)

# Fourier 변환 수행
freq_domain = np.fft.fftshift(np.fft.fftfreq(len(t), t[1] - t[0]))
rect_spectrum = fourier_transform(rect_waveform, t)

# 그래프 그리기
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(t, rect_waveform, label='사각형 펄스')
plt.xlabel('시간')
plt.ylabel('값')
plt.legend()
plt.grid()
plt.title('사각형 펄스 (시간 도메인)')

plt.subplot(2, 1, 2)
plt.plot(freq_domain, np.abs(rect_spectrum), label='사각형 펄스 스펙트럼')
plt.xlabel('주파수')
plt.ylabel('크기')
plt.legend()
plt.grid()
plt.title('사각형 펄스 스펙트럼 (주파수 도메인)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 사각형 펄스 함수 정의
def rect(t, T, A):
    return np.where((t >= -T/2) & (t <= T/2), A, 0)

# Fourier 변환 함수 정의
def fourier_transform(signal, time):
    return np.fft.fftshift(np.fft.fft(signal)) / len(time)

# 시간 범위 설정
t = np.linspace(-5, 5, 1000)  # -5부터 5까지 1000개의 점

# 파라미터 설정
T = 2.0  # 사각형 펄스 주기
A = 1.0  # 사각형 펄스 높이

# 사각형 펄스 계산
rect_waveform = rect(t, T, A)

# Fourier 변환 수행
freq_domain = np.fft.fftshift(np.fft.fftfreq(len(t), t[1] - t[0]))
rect_spectrum = fourier_transform(rect_waveform, t)

# 크기와 위상 스펙트럼 얻기
magnitude_spectrum = np.abs(rect_spectrum)
phase_spectrum = np.angle(rect_spectrum)

# 그래프 그리기
plt.figure(figsize=(12, 6))
plt.subplot(3, 1, 1)
plt.plot(t, rect_waveform, label='사각형 펄스')
plt.xlabel('시간')
plt.ylabel('값')
plt.legend()
plt.grid()
plt.title('사각형 펄스 (시간 도메인)')

plt.subplot(3, 1, 2)
plt.plot(freq_domain, magnitude_spectrum, label='크기 스펙트럼')
plt.xlabel('주파수')
plt.ylabel('크기')
plt.legend()
plt.grid()
plt.title('크기 스펙트럼 (주파수 도메인)')

plt.subplot(3, 1, 3)
plt.plot(freq_domain, phase_spectrum, label='위상 스펙트럼')
plt.xlabel('주파수')
plt.ylabel('위상')
plt.legend()
plt.grid()
plt.title('위상 스펙트럼 (주파수 도메인)')

plt.tight_layout()
plt.show()
```
