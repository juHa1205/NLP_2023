```python
import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(0,2,100)
```

```python
g = 9.81
v0_c = 10
v0_m = 15
```

```python
y_c = -1/2 * g* t**2 + v0_c*t
y_m = -1/2 * g* t**2 + v0_m*t
```

```python
plt.plot(t,y_c, label = 'chulsoo')
plt.plot(t,y_m, label = 'minkoo')
plt.xlabel('Time')
plt.ylabel('Height of the Ball')
plt.grid()
plt.legend()
```
