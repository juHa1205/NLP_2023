```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
X=-2*np.random.rand(100,2)
```

```python
print(X.shape)
X
```

```python
X.max(), X.min, X.std()
```

```python
import seaborn as sns
```

```python
sns.distplot(X[:,0], rug=True)
```

```python
X1 = 1 + 2*np.random.rand(100,2)
sns.distplot(X1[:,0], rug=True)
```

```python
sns.distplot(X1[:,1],rug=True)
```

```python
plt.scatter(X[:,0],X[:,1],s=50,c='b')
plt.show()
```

```python
plt.scatter(X1[:,0],X1[:,1],s=50,c='b')
plt.show()
```

```python
from sklearn.cluster import KMeans
Kmean = KMeans(n_clusters=1)
Kmean.fit(X)
```

```python
X_c=Kmean.cluster_centers_
```

```python
X_c
```

```python
X_c[0]
```

```python
plt.scatter(X[:,0],X[:,1], s=50,c='b')
plt.scatter(X_c[0][0],X_c[0][1],s=200,c='g',marker='s')

plt.show()
```

```python
from sklearn.cluster import KMeans
Kmean = KMeans(n_clusters=1)
Kmean.fit(X1)
```

```python
X1_c=Kmean.cluster_centers_
X1_c
```

```python
plt.scatter(X1[:,0],X1[:,1], s=50,c='b')
plt.scatter(X1_c[0][0],X1_c[0][1],s=200,c='g',marker='s')

plt.show()
```

```python
plt.scatter(X[:,0],X[:,1], s=50,c='b')
plt.scatter(X_c[0][0],X_c[0][1],s=200,c='g',marker='s')
plt.scatter(X1[:,0],X1[:,1], s=50,c='r')
plt.scatter(X1_c[0][0],X1_c[0][1],s=200,c='y',marker='s')
plt.show()
```
